# Download Chrome Driver

From: https://chromedriver.chromium.org/

```python
from selenium import webdriver

from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome(r'C:\Users\Acer\Desktop\WebScrapping\Selenium\chromedriver.exe')
```

```python
driver.get('https://serebii.net/pokemon/gen1pokemon.shtml')
```

```python
from bs4 import BeautifulSoup
```

```python
html = BeautifulSoup(driver.page_source , 'html.parser')
```

```python
pokemons = html.select('table[class="dextable"] tbody tr')
pokemons
```

```python
pokemons[50]
```

```python
driver.get('https://www.random.org/integers')
```

```python
num_numbers = driver.find_element(By.CSS_SELECTOR, "input[name='num']")
num_numbers
```

```python
num_numbers.clear()
```

```python
num_numbers.send_keys('8')
```

```python
submit_button = driver.find_element(By.CSS_SELECTOR, 'input[value="Get Numbers"]')
```

```python
submit_button.click()
```
