## Now lets look what we get from url

```python
data.keys()         # Data dictionary have these keys. Products is most important data for this example.
```

```python
products = data['products']         # We get only products from data.
products
```

## We create empty lists to save data.

```python
titles = []
prices = []
discountPercentages = []
stocks = []
```

## We iterate and get everything we want from products and save them to our lists.

```python
for product in products :
    titles.append(product['title'])
    prices.append(product['price'])
    discountPercentages.append(product['discountPercentage'])
    stocks.append(product['stock'])
```

## Lets create our DataFrame

```python
df = pd.DataFrame({'Title':titles , 'Price':prices , 'DiscountPercentage':discountPercentages , 'Stock':stocks})
df
```
